  <h4>Python List vs NumPy Array - </h4>
  <ul>
    <li><b>Python List:</b> Basic tool (slower, limited operations). List elements are stored separately in memory</li>
    <li><b>NumPy Array:</b> Advanced tool (faster, optimized, more powerful). NumPy stores data in one continuous block of memory</li>
  </ul>

  <h4>Why Use NumPy?</h4>
  <ul>
    <li>Faster than normal Python lists</li>
    <li>For working with multi-dimensional array</li>
    <li>For handling large datasets</li>

<p><b>Copy vs View in NumPy</b> - <br>
<b>view</b> - A view does NOT create new data. It just looks at the original array. If we change view → original array also changes.<br>
<b>Copy</b> - A copy creates a completely new array. If we change copy → original array does NOT change.</p>

````python
arr = np.array([1,2,3,4,5])
view_arr = arr.view()  # Create a view (shares same memory, no new data created)
print(view_arr)   # Print view (same as original initially)

# Modify the view
view_arr[0] = 100
print(arr)    # Change is reflected in original array (because they are linked)
print(view_arr.base) # .base points to original array → proves it's a view
````

````python
arr = np.array([1,2,3,4,5])
copy_arr = arr.copy()  # Create a copy (new memory, completely independent)
print(copy_arr)
copy_arr[0] = 100
print(arr)           # Original array remains unchanged (no shared memory)
print(copy_arr)
print(arr.base)      # .base is None → this array owns its memory (not a view)
````

<p><b>Note : Check if Array Owns its Data</b><br>
Copies owns the data, and views does not own the data, but how can we check this? Every NumPy array has the attribute <b>base</b> that returns <code>None</code> if the array owns the data. Otherwise, the <b>base</b>  attribute refers to the original object.</p>

<h5>Array Iteration in NumPy - </h5>
<p>What is Array Iteration?<br>
Array iteration means accessing each element of a NumPy array one by one. In basic Python, we use nested loops for multi-dimensional arrays, which becomes complex for higher dimensions. <br>

Using <b>np.nditer()</b><br>

`np.nditer()` is an efficient iterator that allows element-wise traversal of arrays regardless of their dimensions. It iterates over each scalar value in the array. </p>

````python
arr = np.array([[1,2,3],[2,6,7]])

# Traditional method (nested loops)
# for i in arr:
#     for j in i:
#         print(j)

# Using nditer
for i in np.nditer(arr):
    print(i)

## Similar for any dimension 
````

<p><b>np.ndenumerate()</b> (Index + Value) - <br>
It iterates with index + element. It is best when we need position tracking.<br>
Syntax : <br></p>  

```python
for index, value in np.ndenumerate(arr):
    print(index, value)

````python
arr = np.array([[1,4,5,7], [2,8,9,5]])

for idx, i in np.ndenumerate(arr):
    print(idx, i)
````

````python
## Joining 2D Arrays using NumPy
arr1 = np.array([[1,2,3],[4,5,6]])
arr2 = np.array([[3,4,5],[2,5,7]])

# axis = 1 → column-wise join (adds more columns, moves right)
arr_axis1 = np.concatenate((arr1, arr2), axis=1)

# axis = 0 → row-wise join (adds more rows, moves down)
# default axis = 0
arr_axis0 = np.concatenate((arr1, arr2))

print(arr_axis1)
print(arr_axis0)
````

<p>To create our own ufunc, we have to define a function, like we do with normal functions in Python, then we add it to our NumPy ufunc library with the <b>frompyfunc()</b> method.<br>

function - the name of the function.<br>
inputs - the number of input arguments (arrays).<br>
outputs - the number of output arrays.</p>

````python
arr1 = np.array([1, 2])
arr2 = np.array([5, 6])
x = np.prod([arr1, arr2])
print(x)
````

<p>If we specify <b>axis=1</b>, NumPy will return the product of each array</p>

````python
arr1 = np.array([1, 2])
arr2 = np.array([5, 6])
x = np.prod([arr1, arr2], axis = 1)
print(x)
````

<p>Finding LCM (Lowest Common Multiple) we use <b>lcm()</b></p>

````python
num1 = 4
num2 = 6
x = np.lcm(num1, num2)
print(x)
````

<p>To find the Lowest Common Multiple of all values in an array, we can use the <b>reduce()</b> method</p>

````python
arr = np.array([3, 6, 9])
x = np.lcm.reduce(arr)
print(x)
````

<p>Finding GCD (Greatest Common Divisor) we use <b>gcd()</b></p>

````python
num1 = 6
num2 = 9
x = np.gcd(num1, num2)
print(x)
````

<p>To find the Highest Common Factor of all values in an array, we can use the <b>reduce()</b> method</p>

````python
arr = np.array([20, 8])
x = np.gcd.reduce(arr)
print(x)
````

<p>We can use NumPy's unique() method to find unique elements from any array. But remember that the set arrays should only be 1-D arrays </p>

````python
arr = np.array([1, 1, 1, 2, 3, 4, 5, 5, 6, 7])
x = np.unique(arr)
print(x)
````

<p>Finding Union - To find the unique values of two arrays, use the <b>union1d()</b> method</p>

````python
arr1 = np.array([1, 2, 3])
arr2 = np.array([5,6])
x = np.union1d(arr1, arr2)
print(x)
````

<p>Finding Intersection - To find only the values that are present in both arrays, use the <b>intersect1d()</b> method</p>

````python
arr1 = np.array([1, 2, 3])
arr2 = np.array([3,5,6])
x = np.intersect1d(arr1, arr2, assume_unique=True)
print(x)
````

<p><b>Note</b>: the <b>intersect1d()</b> method takes an optional argument <b>assume_unique</b>, which if set to <b>True</b> can speed up computation. It should always be set to True when dealing with sets</p>

<p>Finding Difference - To find only the values in the first set that is NOT present in the seconds set, use the <b>setdiff1d()</b> method</p>

````python
set1 = np.array([1, 2, 3, 4])
set2 = np.array([3, 4, 5, 6])
newarr = np.setdiff1d(set1, set2, assume_unique=True)
print(newarr)
````
