  <h4>What is NumPy?</h4>
  <p>
    <b>NumPy (Numerical Python)</b> is a powerful Python library used for 
    fast mathematical computations and working with arrays. It allows us to handle large datasets efficiently and perform complex calculations with minimal code.
  </p>

  <h4>Python List vs NumPy Array - </h4>
  <ul>
    <li><b>Python List:</b> Basic tool (slower, limited operations)</li>
    <li><b>NumPy Array:</b> Advanced tool (faster, optimized, more powerful)</li>
  </ul>

  <h4>Why Use NumPy?</h4>
  <ul>
    <li>Faster than normal Python lists</li>
    <li>For working with multi-dimensional array</li>
    <li>For handling large datasets</li>
    <li>Less code for complex mathematical operations</li>
    <li>Widely used in Data Science, AI, and Machine Learning</li>
  </ul>

  <h4>Key Features of NumPy Array : </h4>
  <ul>
    <li>All elements are usually of the same data type</li>
    <li>Supports vectorized operations (no need for loops)</li>
    <li>Efficient memory usage</li>
  </ul>

<p>Creating arrays - The array object in NumPy is called ndarray. We can create a NumPy ndarray object by using the <b>array()</b> function.</p>

````python
arr = np.array([2,6,7,8])   # it's a 1-D array
print(arr)
print(type(arr))
````

````python
# create a 0-D array
arr = np.array(29)
print(arr)
````

````python
# creating 2-D array
arr = np.array([[1,4,5],[3,6,8]])
print(arr)
````

<p><b>ndim</b> attribute that returns an integer that tells us how many dimensions the array have.</p>

````python
# create 3-D array
arr = np.array([[[1,3,4],[6,7,9]], [[3,1,2],[8,7,5]]])
print(arr)
print("Number of dimension: ",arr.ndim) 
````

<p>Create an array with 5 dimensions. We can define the number of dimensions by using the <b>ndmin</b> argument.</p>

````python
arr = np.array([1,3,4,5] , ndmin = 5)
print(arr)
````

<p><b>Zeros</b> - creates array filled with 0. <br>
Syntax : np.zeros(shape, dtype=float). Here, shape → tuple (rows, cols, …). dtype → optional (default = float). </p>


````python
arr = np.zeros((2,3)) # Here Shape = (2 rows, 3 columns)
print(arr)
## Example of custom data types
# arr = np.zeros((2,2), dtype=int)
# print(arr)
````

<p><b>Ones</b> - creates array filled with 1. The rest same as zeros</p>

````python
arr = np.ones((2,2))
print(arr)
````

<p><b>Full</b> - creates an array of a given shape filled with a specific value. Syntax : np.full(shape, fill_value, dtype=None)

````python
arr = np.full((2,3),7)
print(arr)
````

<p><b>Identity matrix</b> - <br>Syntax : np.identity(n, dtype=None). here, n → number of rows = columns (must be square)</p>

````python
arr = np.identity(3)
print(arr)
````

<p>Custom <b>diagonal matrix</b> - </p>

````python
arr = np.diag([5,5,2])
print(arr)
````

<p><b>np.random.rand()</b> generates an array of random numbers uniformly distributed between 0 and 1.<br>
syntax - np.random.rand(d0, d1, d2, ...)</p>

````python
arr = np.random.rand(2,3)  # 2D array
print(arr)  
````

<p><b>np.random.randint()</b> generates random integers within a specified range.<br>
syntax - np.random.randint(low, high=None, size=None, dtype=int) </p>

````python
x = np.random.randint(5)   # single random num. output - (0-4)
print(x)
y = np.random.randint(10, 20) # range (low to high). output - (10-19)
print(y)

## 1D array
# arr = np.random.randint(1, 10, size=5)
# print(arr)   

## 2D array
arr = np.random.randint(3,29, size = (2,3))
print(arr)
````

<p><b>np.arange()</b> creates a NumPy array with evenly spaced values within a given interval. It’s similar to Python’s built-in range(), but returns a NumPy array.<br>
Syntax - np.arange(start, stop, step, dtype=None).<br>
here, <br>start → starting value (default = 0)<br>stop → ending value (excluded)<br>step → gap between values (default = 1)<br>dtype → optional

````python
arr = np.arange(2,9,2)
print(arr)
````

<p><b>np.linspace()</b> creates an array of evenly spaced numbers between two values, based on the number of elements, not step size.<br>Syntax - np.linspace(start, stop, num=50, endpoint=True, dtype=None).<br>here,<br>start → starting value<br>stop → ending value<br>num → number of elements (default = 50)<br>endpoint → include stop or not<br>dtype → optional</p>

````python
arr = np.linspace(0,60,10)
print(arr)
````

Explanation:

Total elements = 10
Includes both 0 and 60
Automatically calculates step

````python
# Default Behavior

arr = np.linspace(0,1)
print(arr)   # Creates 50 values between 0 and 1
````

````python
# Excluding Endpoint
arr = np.linspace(0, 10, 5, endpoint=False)
print(arr)  # 10 is NOT included
````

````python
# Integer dtype
arr = np.linspace(1, 5, 4, dtype=int)
print(arr)
````

<b>Note: </b>Instead of giving step, NumPy computes:<br>
step = (stop - start) / (num - 1)<br>
(if endpoint=True)

<p>- Array Slicing -<p>

````python
# 1D array 
arr = np.array([2,5,6,7,8])
print(arr[0])
print(arr[1:3])
print(arr[1:])
````

````python
# 2D array
arr = np.array([[1,5,4], [4,7,9]])
print(arr[1,1])  # Think like a matrix. Row 1, Col 1
print(arr[1,:])  # row
print(arr[:,2])  # col
````

<p>max(), min(), argmax(), argmin()<p>

````python
arr = np.array([2,7,8,9,1])
print(arr.max())
print(arr.min())
print(arr.argmax())  # Index of max value (return index)
print(arr.argmin())  # Index of min value (return index)
print(len(arr))
````

<p><b>shape</b> - It tells or sets the structure of array. It shows (rows, columns). It can also change the array directly.</p> 

````python
arr = np.arange(6)
print(arr) 
arr.shape = (2,3)
print(arr)  # changes original array
````

<p><b>reshape()</b> - It returns a new array with new shape. It does not modify original array</p>

````python
arr = np.arange(0,10,2)
print(arr)
matrix = arr.reshape(5,1)
print(matrix)
print(arr)  # Original still same. 
````

**Note:** for **shape** and **reshape()** -
```bash
# Total elements must match
# Example:
# 6 elements → can be (2,3), (3,2), (1,6)
# Cannot be (4,2)
```
